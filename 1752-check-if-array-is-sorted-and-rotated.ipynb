```python
def check(nums):
  master_l = None
  last = nums[0]
  still_non_decr = True

  for curr in nums[1:]:
    delta = curr - last

    # print(f'{curr=}, {last=}, {delta=}')

    if still_non_decr:
      if delta >= 0:
        pass
      else:
        still_non_decr = False
        master_l = nums[0]
        if (curr > master_l):
          return False

    else:  # not still_non_decr
      if delta >= 0 and curr <= master_l:
        pass
      else:
        return False

    last = curr

  return True


# check([1, 3, 2])
# check([2, 4, 1, 2, 4, 4])
# check([3, 4, 5, 1, 2])
# check([2, 1, 3, 4])
# check([1, 2, 3])
check([6, 10, 6])
```
