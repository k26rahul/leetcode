```python
def is_balanced(string):
  brackets = {
      '(': ')',
      '{': '}',
      '[': ']',
  }

  stack = []

  for char in string:
    # char is opening bracket, push it on stack
    if char in brackets:
      stack.append(char)
      continue

    # char is closing bracket, and do we have matching opening bracket on stack?
    if stack and brackets[stack.pop()] == char:
      continue

    return False

  # stack should be empty
  if stack:
    return False

  return True


is_balanced("(")
is_balanced("(]")
is_balanced("]()")

is_balanced("()")
is_balanced("([])")
is_balanced("()[]({{}})")
```
