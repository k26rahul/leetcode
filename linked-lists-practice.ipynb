```python
class Node:
  def __init__(self, val):
    self.val = val
    self.next = None

  def __repr__(self):
    return f'<Node val={self.val}>'


node = Node(7)
print(node)
print(node.next)
```

```python
def lst_to_ll(lst):
  head = None
  tail = None

  for num in lst:
    new_node = Node(num)
    if tail:
      tail.next = new_node
    else:
      head = new_node
    tail = new_node

  return head


def ll_to_lst(head):
  lst = []

  curr = head
  while curr:
    lst.append(curr.val)
    curr = curr.next

  return lst
```

```python
class LL:
  def __init__(self, lst=[]):
    self.head = None
    self.tail = None

    for num in lst:
      self.append(num)

  def append(self, val):
    new_node = Node(val)
    if self.tail:
      self.tail.next = new_node
      self.tail = new_node
    else:
      self.head = new_node
      self.tail = new_node

  def prepend(self, val):
    if not self.head:
      self.append(val)
      return

    new_node = Node(val)
    new_node.next = self.head
    self.head = new_node

  def __repr__(self):
    lst = []
    curr = self.head
    while curr:
      lst.append(curr.val)
      curr = curr.next
    return str(lst)


ll = LL()
ll = LL([1, 2, 3])
ll.prepend(5)
print(ll)
```
