```python
def check(nums):
  n = len(nums)

  pl = []
  pr = []

  pl_c = 1
  pr_c = 1
  for i in range(len(nums)):
    pl_c *= nums[i]
    pl.append(pl_c)

    pr_c *= nums[(n-1)-i]
    pr.append(pr_c)

  # print(pl)
  # print(pr)

  ans = []
  for i in range(n):
    term_l = 1 if i == 0 else pl[i-1]
    term_r = 1 if i == n-1 else pr[(n-1)-1-i]
    # print(f'{i=}, {term_l=}, {term_r=}')
    ans.append(term_l * term_r)

  return ans


check([1, 2, 3, 4])
check([-1, 1, 0, -3, 3])
```
