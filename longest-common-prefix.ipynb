```python
def longest_common_prefix(words):
  candidates = {}

  first_word = words[0]
  for i in range(1, len(first_word)+1):
    candidates[i] = first_word[:i]

  last_length = len(candidates)

  for word in words:
    # print(f'dealing with {word=}')

    if word == '':
      return ''

    last_length = min(last_length, len(word))

    for i in range(1, last_length+1):
      w_prefix = word[:i]
      c_prefix = candidates[i]
      # print(f'  {w_prefix}, {c_prefix}')

      if w_prefix != c_prefix:
        last_length = i-1
        # print(f'   ‚ö†Ô∏è {last_length=}')

        if last_length == 0:
          # print(f'    üíÄ no common prefix possible')
          return ''

        break

  return candidates[last_length]
```

```python
longest_common_prefix(["flower", "flow", "flight"])
longest_common_prefix(["flower", "flow", ""])
longest_common_prefix(["flower", "flow"])
longest_common_prefix(["a", "b"])
```

```python
def longest_common_prefix(words):
  ans = words[0]

  for word in words[1:]:
    # print(f'dealing with {word=}')

    if len(word) == 0:
      return ''

    if len(word) < len(ans):
      ans = ans[:len(word)]
      # print(f'    ‚ö†Ô∏è1 {ans=}')

    for i in range(1, min(len(word)+1, len(ans)+1)):
      w_prefix = word[:i]
      ans_prefix = ans[:i]
      # print(f'    {w_prefix}, {ans_prefix}')

      if w_prefix != ans_prefix:
        ans = ans[:i-1]
        # print(f'    ‚ö†Ô∏è2 {ans=}')
        break

  return ans
```

```python
longest_common_prefix(["flower", "flow", "flight"])
longest_common_prefix(["flower", "flow", ""])
longest_common_prefix(["flower", "flow"])
longest_common_prefix(["a", "b"])
```
