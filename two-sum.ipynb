```python
nums = [11, 15, 2, 7, 23, 16]
target = 9
```

```python
# O(n^2)
for i in range(len(nums)):  # O(n)
  print(f'dealing with {nums[i]}')

  for j in range(i+1, len(nums)):  # O(n)
    print(f'  possible partner: {nums[j]}')

    if nums[i] + nums[j] == target:
      print(f'  üòç ans found: {i, j}')
      break
```

```python
# better solution, O(2n)
num_to_index = {}

for i, num in enumerate(nums):  # prepare the map, O(n)
  num_to_index[num] = i

print(f'map is: {num_to_index}')

for i, num in enumerate(nums):  # enjoy the map, O(n)
  partner = target - num

  print(f'dealing with {num=}, {partner=}')

  if partner in num_to_index:
    print(f'  üòç partner found, ans: {(i, num_to_index[partner])}')
    break
```

```python
def two_sum(nums, target):
  # prepare the map, O(n)
  num_to_index = {}
  for i, num in enumerate(nums):
    num_to_index[num] = i

  # print(f'map is ready: {num_to_index}')

  # enjoy the map, O(1)
  for i, num in enumerate(nums):
    partner = target - num
    # print(f'dealing with {num=}, {partner=}')

    if partner in num_to_index:  # magic: O(1), can check if partner exist
      partner_idx = num_to_index[partner]
      if partner_idx != i:
        # print(' üòç partner found')
        return (i, partner_idx)
```

```python
nums = [3, 2, 4]
target = 6

two_sum(nums, target)
```
