```python
def check(board):
  for row in board:
    found = {}
    for num in row:
      if num == '.':
        continue
      if num in found:
        return False
      found[num] = True

  for j in range(9):
    found = {}
    for i in range(9):
      num = board[i][j]
      if num == '.':
        continue
      if num in found:
        return False
      found[num] = True

  for i in range(0, 9, 3):
    for j in range(0, 9, 3):
      found = {}
      for i2 in range(i, i+3):
        for j2 in range(j, j+3):
          num = board[i2][j2]
          if num == '.':
            continue
          if num in found:
            return False
          found[num] = True

  return True


board = [["5", "3", ".", ".", "7", ".", ".", ".", "."],
         ["6", ".", ".", "1", "9", "5", ".", ".", "."],
         [".", "9", "8", ".", ".", ".", ".", "6", "."],
         ["8", ".", ".", ".", "6", ".", ".", ".", "3"],
         ["4", ".", ".", "8", ".", "3", ".", ".", "1"],
         ["7", ".", ".", ".", "2", ".", ".", ".", "6"],
         [".", "6", ".", ".", ".", ".", "2", "8", "."],
         [".", ".", ".", "4", "1", "9", ".", ".", "5"],
         [".", ".", ".", ".", "8", ".", ".", "7", "9"]]
check(board)
```
