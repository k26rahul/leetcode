```python
symbol_to_value = {
    'I': 1,
    'V': 5,
    'X': 10,
    'L': 50,
    'C': 100,
    'D': 500,
    'M': 1000,
}
```

```python
def roman_to_integer(input_str):
  ans = 0
  last_value = float('inf')

  for symbol in input_str:
    print(f'dealing with {symbol=}')

    value = symbol_to_value[symbol]
    if value <= last_value:
      ans += value
      print(f'  +{value}')
    else:
      ans -= last_value
      print(f'  -{value}')

      ans += (value - last_value)
      print(f'  +{value - last_value}')

    last_value = value

  return ans


roman_to_integer('CM')
```
