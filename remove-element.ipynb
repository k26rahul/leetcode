```python
nums = [3, 2, 2, 3]
val = 3

# Extract all elements NOT equal to value
not_equals = [num for num in nums if num != val]
not_equals

k = len(not_equals)
nums[:k] = not_equals  # In-place replace nums

# Done. First, k elements NOT equal to value
# Also we don't care what comes after k-th element
nums
```

```python
# In-place replace
fruits = ['apple', 'banana', 'kiwi', 'mango']
fruits[:2] = ['orange', 'lemon']
fruits
```

```python
def remove_element(nums):
  # Extract all elements NOT equal to value
  not_equals = [num for num in nums if num != val]
  k = len(not_equals)

  # In-place replace nums
  nums[:k] = not_equals
  return k


nums = [3, 2, 2, 3]
val = 3
remove_element(nums)
nums
```
