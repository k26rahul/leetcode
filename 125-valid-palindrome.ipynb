```python
s = "A man, a plan, a canal: Panama"
new_s = []

for char in s:
  o = ord(char)
  is_upper = 65 <= o <= 90
  is_lower = 97 <= o <= 122
  is_digit = 48 <= o <= 57
  is_alpha = is_upper or is_lower or is_digit

  # print(f'{char=}, {o=}, {is_upper=}, {is_lower=}, {is_digit=}')

  if is_upper:
    o = 97 - 65 + o

  if is_alpha:
    new_s.append(o)

print(new_s)
# print([chr(o) for o in new_s])

new_s == new_s[::-1]
```

```python
s = "A man, a plan, a canal: Panama"


def helper(char):
  o = ord(char)
  is_upper = 65 <= o <= 90
  is_lower = 97 <= o <= 122
  is_digit = 48 <= o <= 57
  is_alpha = is_upper or is_lower or is_digit

  # print(f'{char=}, {o=}, {is_upper=}, {is_lower=}, {is_digit=}')

  if is_upper:
    o = 97 - 65 + o

  return o, is_alpha


for i in range(len(s)):
  helper(s[i])
```
